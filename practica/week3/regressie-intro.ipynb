```python
from numpy import array, exp
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
x = 25000
y = 5000
h = np.arange(50, 200)
z = x + y*h
c = np.random.normal(h, 6)

```

```python
y = np.array(z) 

x = np.array(c)

def func1(a, x, b):
    return a*x+b

fit, var = curve_fit(func1, x, y)
print("fit: ", fit)

fit, _ = curve_fit(func1, x, y)
a, b = fit[0], fit[1]
yfit1 = a*x+b

plt.plot(x, y, 'bo', label="y=noisy house price")
plt.plot(x, yfit1, label="y=original house price")

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```

### Stap 1: Maak model bolmannetje

```python
# ...
```

```python
benen = 12.4
armen = 8.7
hoofd = 3.2
radius = np.arange(1, 10, 0.1)
r = np.random.normal(mass_bolman, 20.3)
mass_bolman = 1.03 * (radius**3) + benen + armen + hoofd

# mass_bolman = 1.03*r**3 + benen + armen + hoofd
plot_graph(radius, r)
```

```python
y = np.array(r) 

x = np.array(mass_bolman)

def func2(a, x, b):
    return a*(x**3)+b

fit, var = curve_fit(func2, x, y)

fit, _ = curve_fit(func2, x, y)
a, b = fit[0], fit[1]
yfit1 = a*x**3+b

plt.plot(x, y, 'bo', label="y=noisy bolman weigth")
plt.plot(x, yfit1, label="y=original bolman weigth")

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```
